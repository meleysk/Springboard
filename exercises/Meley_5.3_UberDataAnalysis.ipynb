```python
%pylab inline
import pandas
import seaborn #seaborn makes pretty plots without much effort
```

## Load CSV file into memory

```python

data = pandas.read_csv('uber-raw-data-apr14.csv')
data.head()
```

## Convert datetime and add some useful columns


```python
#example of how datetime works
dt ='4/1/2014 0:11:00'
dt = pandas.to_datetime(dt)
dt.day
dt.month
```

```python
data['Date/Time'] = data['Date/Time'].map(pandas.to_datetime)
```

```python
data.tail()
```

```python
def get_dom(dt):
    return dt.day

data['dom'] = data['Date/Time'].map(get_dom)
    
```

```python
data.tail()
```

```python
def get_weekday(dt):
    return dt.weekday()

data['weekday'] = data['Date/Time'].map(get_weekday)

def get_hour(dt):
    return dt.hour

data['hour']= data['Date/Time'].map(get_hour)
data.tail()
```

# analysis

## analyze the DoM

```python
hist(data.dom , bins =31, rwidth =.8, range=(0.5,30.5))
xlabel('date of the month')
ylabel('frequency')
title('Frequency by DoM - uber - Apr 2014')
```

```python
# for k, rows in data.groupby('dom'):
#     print((k,len(rows)))

#this is the pandas version of the above for loop

def count_rows(rows):
    return len(rows)

by_date = data.groupby('dom').apply(count_rows)
by_date
```

```python
#now lets go ahead and plot our new by_date data

bar(range(1,31), by_date)
```

```python
by_date_sorted = by_date.sort_values()
by_date_sorted
```

```python
bar(range(1,31),by_date_sorted)
xticks(range(1,31), by_date_sorted.index)
xlabel('date of the month')
ylabel('frequency')
title('Frequency by DoM - uber - Apr 2014');
```

## analize the hour

```python
hist(data.hour, bins=24, range =(.5,24))
```

## analyze the weekday 

```python
#hist(data.weekday, bins=7, range = (-.5,7), rwidth = .8, color = 'green')
hist(data.weekday, bins=7, range = (-.5,7), rwidth = .8, color = '#AA6666', alpha =.5)
xticks(range(7), 'Mon Tue Wed Thu Fri Sat Sun'.split())
xlabel('day of week')
ylabel('frequency')
title('Frequency by day of week - uber - Apr 2014');
```

## cross analysis(hour,dow)

```python
#unstuck() gives you the option to have hour as a column name and weekday as a row
by_cross = data.groupby('weekday hour'.split()).apply(count_rows).unstack()
by_cross
```

```python
#now we can use a heat map
seaborn.heatmap(by_cross);
```

we can see from the heatmap that uber requests are higher after 3 PM and Friday and saturday
people are staying out late. we can also see that sunday morning people are coming home early
